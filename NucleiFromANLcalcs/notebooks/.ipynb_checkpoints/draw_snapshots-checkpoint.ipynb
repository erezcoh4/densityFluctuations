```python
def draw_nucleus( X , Y , Z ):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111, projection='3d')
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x = 10 * np.outer(np.cos(u), np.sin(v))
    y = 10 * np.outer(np.sin(u), np.sin(v))
    z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))

    for i in range(5):
        ax.plot_surface(x + X[i], y + Y[i], z + Z[i], rstride=4, cstride=4, color='blue' , alpha=0.25)
    
    for i in range(6,10):
        ax.plot_surface(x + X[i], y + Y[i], z + Z[i], rstride=4, cstride=4, color='red' , alpha=0.25)
        
    plt.show()
```

```python
draw_nucleus( Nx[2] , Ny[2] , Nz[2] )
```
